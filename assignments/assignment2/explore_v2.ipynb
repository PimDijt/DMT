```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction import DictVectorizer

with open ('feature_dict', 'rb') as fp:
    training_data = pickle.load(fp)
    
training_vec = DictVectorizer()
training_data = training_vec.fit_transform(training_data).toarray()
```

```python
dataframe = pd.DataFrame(training_data)

plt.matshow(dataframe.corr())
```

```python
dataframe.corr()
```

```python
dataframe = (dataframe-dataframe.mean())/dataframe.std(ddof=0)
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(dataframe)
```

```python
plt.figure(1, figsize=(4, 3))
plt.clf()
plt.axes([.2, .2, .7, .7])
plt.plot(pca.explained_variance_, linewidth=2)
plt.axis('tight')
plt.xlabel('n_components')
plt.ylabel('explained_variance_')
```
